```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload file (kalau di Colab)
from google.colab import files
uploaded = files.upload()

# Ambil nama file
file_name = list(uploaded.keys())[0]

# Baca citra grayscale
img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

# =========================
# Transformasi Negatif
# =========================
negatif = 255 - img

# =========================
# Transformasi Logaritmik
# =========================
log_img = np.log(1 + img.astype(np.float32))

# Normalisasi ke 0–255
log_img = log_img / np.max(log_img) * 255

log_img = np.array(log_img, dtype=np.uint8)

# =========================
# Histogram
# =========================
def plot_histogram(image, title):
    plt.hist(image.ravel(), bins=256, range=[0,256])
    plt.title(title)
    plt.xlabel("Intensitas")
    plt.ylabel("Frekuensi")

# =========================
# Tampilkan hasil
# =========================
plt.figure(figsize=(12,10))

# Citra asli
plt.subplot(3,2,1)
plt.imshow(img, cmap='gray')
plt.title("Citra Asli")
plt.axis('off')

# Histogram asli
plt.subplot(3,2,2)
plot_histogram(img, "Histogram Asli")

# Negatif
plt.subplot(3,2,3)
plt.imshow(negatif, cmap='gray')
plt.title("Citra Negatif")
plt.axis('off')

# Histogram negatif
plt.subplot(3,2,4)
plot_histogram(negatif, "Histogram Negatif")

# Logaritmik
plt.subplot(3,2,5)
plt.imshow(log_img, cmap='gray')
plt.title("Citra Logaritmik")
plt.axis('off')

# Histogram log
plt.subplot(3,2,6)
plot_histogram(log_img, "Histogram Logaritmik")

plt.tight_layout()
plt.show()
```
